```python
import matplotlib.pyplot as plt
import numpy as np

# Load the metrics from the .npz file
metrics_file = 'metrics.npz'
data = np.load(metrics_file)

# Extract train and validation accuracies
train_accuracies = data['train_accuracies']
valid_accuracies = data['valid_accuracies']

# Plot the train and validation accuracies
plt.figure(figsize=(10, 6))
plt.plot(train_accuracies, label='Train Accuracy')
plt.plot(valid_accuracies, label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Train and Validation Accuracy over Epochs')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Extract train and validation accuracies
train_accuracies = data['train_losses']
valid_accuracies = data['valid_losses']

# Plot the train and validation accuracies
plt.figure(figsize=(10, 6))
plt.plot(train_accuracies, label='Train Loss')
plt.plot(valid_accuracies, label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Dendrites Train and Validation Loss over Epochs')
plt.legend()
plt.grid(True)
plt.savefig('dendrite.png')
plt.show()
```
